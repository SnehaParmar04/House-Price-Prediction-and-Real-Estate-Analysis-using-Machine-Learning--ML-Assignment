```python

DATA_PATH = "/kaggle/input/datasets/alyelbadry/house-pricing-dataset/house_prices.csv"
import pandas as pd, matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.linear_model import LinearRegression

df=pd.read_csv(DATA_PATH).dropna()
X=pd.get_dummies(df.drop("price",axis=1))
y=df["price"]

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)

m=LinearRegression()
m.fit(X_train,y_train)
p=m.predict(X_test)

table=pd.DataFrame([["Linear",mean_absolute_error(y_test,p),r2_score(y_test,p)]],columns=["Model","MAE","R2"])
table.to_csv("RQ5_table.csv",index=False)

plt.figure()
plt.plot(["MAE","R2"],table.iloc[0][1:],marker='o')
plt.savefig("RQ5_figure.pdf")
plt.show()
```
