```python

DATA_PATH = "/kaggle/input/datasets/alyelbadry/house-pricing-dataset/house_prices.csv"
import pandas as pd, matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

df = pd.read_csv(DATA_PATH).dropna()
X = pd.get_dummies(df.drop("price",axis=1))
y = df["price"]

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)

m=RandomForestRegressor()
m.fit(X_train,y_train)
mae=mean_absolute_error(y_test,m.predict(X_test))

table=pd.DataFrame([["Processed",mae]],columns=["Stage","MAE"])
table.to_csv("RQ3_table.csv",index=False)

plt.figure()
plt.plot(table["Stage"],table["MAE"],marker='o')
plt.savefig("RQ3_figure.pdf")
plt.show()
```
